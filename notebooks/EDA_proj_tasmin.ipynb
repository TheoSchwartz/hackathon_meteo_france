```python
import xarray as xr
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
def get_filename(file:str)->str:
    filename = file.rsplit('.nc')[0]
    filename = filename.rsplit('/')[-1]
    return filename


def get_year_range(filename:str):
    years = filename.rsplit('_')[-1]
    year_start, year_end = years.split('-')
    year_range = range(int(year_start[:4]), int(year_end[:4])+1)
    return year_range


def compute_nc_files(file:str):
    ds = xr.open_dataset(file)
    filename = get_filename(file)
    year_range = get_year_range(filename=filename)
    print(year_range)
    for year in year_range:
        ds_year = ds[['tasminAdjust']].sel(time = slice(f"{year}-06-01", f"{year}-09-30"))
        to_write = ds_year.to_dataframe().dropna()
        to_write.to_parquet(f"../data/projections_parquet/{filename}_{year}.parquet")
        print(f"{filename} saved ✅")

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    dphi = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)
    a = np.sin(dphi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(dlambda/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))
```

```python
lat0 = 48.844304
lon0 = 2.374377
dict_consecutive_nights = {}
list_parquet_files = os.listdir("/Users/emile/Desktop/Datathon_MF/data/projections_parquet/")
for parquet_file in list_parquet_files:
    df_points = pd.read_parquet(f"/Users/emile/Desktop/Datathon_MF/data/projections_parquet/{parquet_file}")
    df_points["distance_km"] = haversine(lat0, lon0,
                                        df_points["lat"],
                                        df_points["lon"])

    nearest = df_points.loc[df_points["distance_km"].idxmin()]
    nearest_lat = nearest["lat"]
    nearest_lon = nearest["lon"]

    df_nearest = df_points[(df_points["lat"] == nearest_lat) & (df_points["lon"] == nearest_lon)]
    df_nearest_two_nights = df_nearest[df_nearest["two_consecutive_nights"] == True]
    print(df_nearest_two_nights.shape)
    year = parquet_file.rsplit('_')[-1].rsplit('.')[0]
    year = year[:4]
    dict_consecutive_nights[year] = df_nearest_two_nights.shape[0]
    print(f"Year: {year}, Number of two consecutive hot nights at nearest point ({nearest_lat}, {nearest_lon}): {df_nearest_two_nights.shape[0]}")
```

```python
import plotly.express as px

# Sort your dict by key
dict_consecutive_nights = dict(sorted(dict_consecutive_nights.items()))

# Convert to lists
years = list(dict_consecutive_nights.keys())
counts = list(dict_consecutive_nights.values())

# Create line plot
fig = px.line(
    x=years,
    y=counts,
    markers=True,   # equivalent to marker='o'
    labels={'x': 'Année', 'y': "Nombre d'occurence de nuits consécutives chaudes (>20°C)"},
    title="Nombre d'occurence de nuits consécutives chaudes (>20°C) Gare de Lyon par année"
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

dict_consecutive_nights = dict(sorted(dict_consecutive_nights.items()))
df = pd.DataFrame({
    "year": list(dict_consecutive_nights.keys()),
    "value": list(dict_consecutive_nights.values())
})

df["decade"] = (df["year"].astype(int) // 10) * 10  + 5  

#moyenne par décennie
df_decade = df.groupby("decade", as_index=False)["value"].mean()
df_decade.rename(columns={"value": "decade_mean"}, inplace=True)

fig = px.line(df, x="year", y="value", markers=True,
              labels={"value": "Occurences de nuits consécutives chaudes (>20°C)"},
              title="Nombre d'occurences de nuits consécutives chaudes (>20°C) Gare de Lyon par année")

fig.add_scatter(
    x=df_decade["decade"],
    y=df_decade["decade_mean"],
    mode="lines+markers",
    name="Moyenne par décennie"
)

fig.show()
```
